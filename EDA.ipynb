```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/Titanic Dataset.csv')
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu", yticklabels=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.drop(columns=['body', 'cabin', 'boat'], inplace=True)
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
df['fare'].fillna(df['fare'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['home.dest'], inplace=True)
```

```python
df.info()
```

```python
sns.countplot(x='sex', data=df)
plt.title("Count of Passengers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='pclass', data=df)
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='embarked', data=df)
plt.title("Passenger Count by Port of Embarkation")
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True, bins=30)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['fare'], kde=True, bins=40)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=df)
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
sns.boxplot(x='survived', y='fare', data=df)
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Now plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python

correlation_matrix = numeric_df.corr()
print(correlation_matrix['survived'].sort_values(ascending=False))
```

```python
sns.pairplot(data=numeric_df, hue='survived', corner=True, palette='Set1')
plt.suptitle("Pairwise Relationships (Colored by Survival)", y=1.02)
plt.show()
```
